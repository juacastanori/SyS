##Potencia circuito RL:
$$L = 0.01 H\\R = 4 \Omega\\i(t) = 2\sin(300t)$$

````python
import matplotlib.pyplot as plt
import numpy as np

w = 300
L = 0.01
R = 4
Imax = 2
T0 = 2*np.pi/w
N = 100 #número de muestras

t = np.linspace(0, T0, N)
i = Imax*np.sin(w*t)

Vmax = np.sqrt(np.square(R)+np.square(w*L))*Imax
theta = np.arctan(w*L/R)
v = Vmax*np.sin(w*t + theta)

plt.figure(figsize=[10,5])
plt.plot(t, i, label = "i(t)")
plt.plot(t, v, label = "v(t)")
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()

p = v*i #potencia instantanea
P = np.trapz(p, t)/T0 #Potencia media TAMBIEN SE PUEDE RESOLVER POR SYMPY
print("Potencia media: "+str(P))
````

````python
plt.figure(figsize=[10,5])
plt.plot(t, p, label = "Potencia instantanea p(t)")
plt.plot(t, P*np.ones(len(t)), "g", label = "Potencia media P")
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
````

````python
#Realizado con sympy
tt= symbols('tt', real = True)
w = 300
L = 0.01
R = 4
Imax = 2
T0 = 2*pi/w
i = Imax*sin(w*tt)
Vmax = sqrt(R**2+(w*L)**2)*Imax
theta = np.arctan(w*L/R)
v = Vmax*sin(w*tt + theta)
pp=v*i
Pv = integrate(pp, (tt, 0, T0)) /T0
print('La potencia media de la señal es:')
Pv


````

##Potencia circuito RC:
$$
R = 4\Omega\\
C = 0.01 F\\
i(t) = 2\sin(300t)
$$

````python
C = 0.01
R=4
w=300
Imax=2
T0 = 2*np.pi/w
N = 100 #número de muestras

t = np.linspace(0, T0, N)
i = Imax*np.sin(w*t)

Vmax = np.sqrt(np.square(R)+np.square(1/(w*C)))*Imax
theta = -np.arctan(1/(w*R*C))
v = Vmax*np.sin(w*t + theta)

plt.figure(figsize=[10,5])
plt.plot(t, i, label = "i(t)")
plt.plot(t, v, label = "v(t)")
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()

p = v*i #potencia instantanea
P = np.trapz(p, t)/T0 #Potencia media TAMBIEN SE PUEDE RESOLVER POR SYMPY
print("Potencia media: "+str(P))
````

````python
plt.figure(figsize=[10,5])
plt.plot(t, p, label = "Potencia instantanea p(t)")
plt.plot(t, P*np.ones(len(t)), "g", label = "Potencia media P")
plt.grid()
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
````

````python
#Realizado con sympy
tt= symbols('tt', real = True)
C = 0.01
R=4
w=300
Imax=2
T0 = 2*pi/w
i = Imax*sin(w*tt)
Vmax = sqrt(R**2+(1/(w*C))**2)*Imax
theta = -np.arctan(1/(w*R*C))
v = Vmax*sin(w*tt + theta)
pp=v*i
Pv = integrate(pp, (tt, 0, T0)) /T0
print('La potencia media de la señal es:')
Pv

````

$$
v(t) = A\cos(w_0t), ~~~ w_0 = \frac{2\pi}{T_0}
$$

Determinar su potencia media y energía total.

$$
\bar{P}_v = \frac{1}{T_0} \int_0^{T_0} |A\cos(w_0t)|^{2}~dt\\
\
$$

El valor RMS de una señal $x(t)=v(t)$ se calcula:

````python
t = symbols('t', real = True)
w, Vmax = symbols('w, Vmax', real = True, nonzero = True, nonnegative = True)
T0 = 2*pi/w
v = Vmax*sin(w*t)
Pv = sqrt(integrate(v**2, (t, 0, T0)) /T0)
print('El valor rms es: ')
Pv
````

El valor de la potencia media es:

````python
Pv = (integrate(v**2, (t, 0, T0)) /T0)
Pv
````
